```python
import numpy as np
```

```python
A = np.identity(10)
A[np.arange(9), np.arange(9)+1] = 1
A
```

```python
A_eval, A_evec = np.linalg.eig(A)
print(A_eval)
print(A_evec)
```

```python
A_tilde = A.copy()
A_tilde[9, 0] = 10**-10
A_tilde
```

```python
A_tilde_eval, A_tilde_evec = np.linalg.eig(A_tilde)
print(A_tilde_eval)
print(A_tilde_evec)
```

```python
diff = np.abs(A_tilde_eval.real - A_eval)
```

```python
diff.mean()
```

```python
p = lambda z: z ** 6 - 1
g = lambda z: (6 * z ** 5) / (z**6 - 1)
# u_prime = lambda z: -(6 * z ** 4 * (z**6 + 5))/((z**6 - 1)**2)
h = lambda z: g(z) ** 2 - (30 * z ** 4) / (z**6 -1)

```

```python
def laguerre(x, g, h, n=6):
    L_p = (g(x) + np.sqrt((n - 1) * (n * h(x) - g(x) ** 2)))
    L_m = (g(x) - np.sqrt((n - 1) * (n * h(x) - g(x) ** 2)))

    if abs(L_p) > abs(L_m):
        return x - n / L_p
    else:
        return x - n / L_m
    
def L(x):
    s = np.sign((x ** 3).real)
    z = s * x ** 3
    return x ** -2 * s * (1 + 5*z)/(z + 5)
```

```python
rand_p = np.apply_along_axis(lambda args: [complex(*args)], 0, np.random.rand(2,100))[0]
print(rand_p)
```

```python
lag = []
an = []
for c in rand_p:
    lag.append(laguerre(c, g, h))
    an.append(L(c))
lag = np.array(lag)
an = np.array(an)
```

```python
(abs((lag - an))**2).mean()
```
