```python
import numpy as np
```

```python
A = np.identity(10)
A[np.arange(9), np.arange(9)+1] = 1
A
```

```python
A_eval, A_evec = np.linalg.eig(A)
print(A_eval)
print(A_evec)
```

```python
A_tilde = A.copy()
A_tilde[9, 0] = 10**-10
A_tilde
```

```python
A_tilde_eval, A_tilde_evec = np.linalg.eig(A_tilde)
print(A_tilde_eval)
print(A_tilde_evec)
```

```python
diff = np.abs(A_tilde_eval.real - A_eval)
```

```python
diff.mean()
```
