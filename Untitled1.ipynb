```python
from sklearn.datasets import make_circles
from sklearn.cluster import SpectralClustering, KMeans, dbscan
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data, labels = make_circles(n_samples=1000, shuffle=True, noise=0.05, random_state=None, factor=0.2)
```

```python
plt.scatter(data[:, 0], data[:, 1])
```

```python
spectral = SpectralClustering(n_clusters=2, affinity="nearest_neighbors", eigen_solver="arpack").fit(data)
kmeans = KMeans(n_clusters=2).fit(data)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=spectral.labels_)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_)
```
